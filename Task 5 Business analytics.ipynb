# EDA

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(),annot=True,cmap='RdBu')
```

```python
sns.set_context("poster", font_scale = 0.6, rc={"grid.linewidth": 5})
sns.set_style('darkgrid')
sns.catplot(x='Region', data=df,kind='count',height=6,aspect=1.5,palette='husl',).set_axis_labels('region of shipping',
                                                                                                      'no. of orders')
```

```python
plt.figure(figsize=(15,7))
sns.set_style('darkgrid')
abc=df['State'].value_counts()[:10]
sns.barplot(x=abc,y=abc.index,palette='Set1')
plt.title("Top ten states (sales wise)",size=20)
plt.xlabel("Number of orders",size=15)
```

```python
plt.figure(figsize=(15,7))
sns.set_style('darkgrid')
abc=df['City'].value_counts()[:10]
sns.barplot(x=abc,y=abc.index,palette='Set1')
plt.title("Top ten cities (sales wise)",size=20)
plt.xlabel("Number of orders",size=15)
```

```python
sns.set_context("poster", font_scale = 0.6, rc={"grid.linewidth": 5})
sns.set_style('darkgrid')
sns.catplot(x='Ship Mode', data=df,kind='count',height=6,aspect=1.5,palette='husl',).set_axis_labels('mode of shipping',
                                                                                                      'no. of orders')
```

```python
sns.set_context("poster", font_scale = 0.6, rc={"grid.linewidth": 5})
sns.set_style('darkgrid')
sns.catplot(x='Segment', data=df,kind='count',height=6,aspect=1.5,palette='husl',).set_axis_labels('Segments',
                                                                                                      'no. of orders')
```

```python
sns.set_context("poster", font_scale = 0.6, rc={"grid.linewidth": 5})
sns.set_style('darkgrid')
sns.catplot(x='Category', data=df,kind='count',height=6,aspect=1.5,
            palette='husl',).set_axis_labels('Category of order','no. of orders')

plt.figure(figsize=(20,8))
values=[6026,2121,1847]
labels= ['Office Supplies','Furniture','Technology']
plt.pie(values,labels=labels,autopct='%1.1f%%',counterclock=True)
plt.title('Categories of order')
plt.show()
```

```python
sns.set_context("poster", font_scale = 0.6, rc={"grid.linewidth": 5})
sns.set_style('darkgrid')
sns.catplot(x='Sub-Category', data=df,kind='count',height=7,aspect=2.5,
            palette='husl',).set_axis_labels('Category of order','no. of orders')
```

```python
plt.figure(figsize=(15,7))
sns.set_style('darkgrid')
abc=df[df['Category']=='Furniture']['Sub-Category'].value_counts()
sns.barplot(x=abc,y=abc.index,palette='Set1')
plt.title("Furniture sales",size=20)
plt.xlabel("Number of orders",size=15)
plt.ylabel("Items",size=15)

plt.figure(figsize=(20,8))
values=[957,617,319,228]
labels= ['Furnishings','Chairs','Tables','Bookcases']
plt.pie(values,labels=labels,autopct='%1.1f%%',counterclock=True)
plt.title('Categories of order')
plt.show()

```
